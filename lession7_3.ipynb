```python
import wget
wget.download('https://github.com/EveYang0630/python_analysis/raw/refs/heads/main/ChineseFont.ttf')
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')
```

```python
fig = plt.figure(figsize=(10, 6))
axes = fig.add_subplot(1,1,1)
axes.set_xticks([1,3,5,7,9])
axes.set_yticks([-1,-0.5,0,0.5,1])

plt.show()
```

```python
#軸tick自動設定
fig = plt.figure(figsize=(10, 6))
axes = fig.add_subplot(1,1,1)
axes.plot([11,3,5,7,9], marker='o', color='red', linestyle='')

plt.show()
```

```python
#畫兩張axes在一個figure
fig = plt.figure(figsize=(10, 15))
axes1 = fig.add_subplot(3,1,1)
axes1.plot([11,3,5,7,9], marker='o', color='red', linestyle='')
axes2 = fig.add_subplot(3,1,2)
axes2.plot([11,3,5,7,9], marker='x', color='g', linestyle='--')
axes3 = fig.add_subplot(3,1,3)
axes3.bar(['A','B','C','D','E'],[11,3,5,7,9])

plt.show()
```
