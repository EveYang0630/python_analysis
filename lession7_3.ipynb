```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')
```

```python
fig = plt.figure(figsize=(10, 6))
axes = fig.add_subplot(1,1,1)
axes.set_xticks([1,3,5,7,9])
axes.set_yticks([-1,-0.5,0,0.5,1])

plt.show()
```

```python
#軸tick自動設定
fig = plt.figure(figsize=(10, 6))
axes = fig.add_subplot(1,1,1)
axes.plot([11,3,5,7,9], marker='o', color='red', linestyle='')

plt.show()
```

```python
#畫兩張axes在一個figure
fig = plt.figure(figsize=(8, 12))
axes1 = fig.add_subplot(3,1,1)
axes1.plot([11,3,5,7,9], marker='o', color='red', linestyle='')
axes2 = fig.add_subplot(3,1,2)
axes2.plot([11,3,5,7,9], marker='x', color='g', linestyle='--')
axes3 = fig.add_subplot(3,1,3)
axes3.bar(['A','B','C','D','E'],[11,3,5,7,9])

plt.show()
```

```python
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
```

```python
subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10, 6))
axes = figure.add_subplot(1,1,1)
axes.set_title('學生成績比較')
axes.set_xlabel('科目')
axes.set_ylabel('分數')
axes.plot(subject,score1,'ro-.',label='學生1')
axes.plot(subject,score2,'gx--',label='學生2')
axes.legend()
axes.grid(axis='y')

plt.show()
```
