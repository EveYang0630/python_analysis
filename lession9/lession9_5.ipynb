```python
# 範例 DataFrame

import pandas as pd
import numpy as np
import matplotlib as mpl

df = pd.DataFrame({
    "strings": ["Adam", "Mike"],
    "ints": [1, 3],
    "floats": [1.123, 1000.23]
})
df
```

- format()
- format_index()

```python
# format()格式化 DataFrame 的數值顯示（小數點後 2 位，千位分隔符等）
# format_index(axis=1,formatter=str.upper)將欄位名稱轉換為大寫。
# relabel_index()將行索引重新命名為 row1 和 row2

df.style\
.format(
    precision=2,
    thousands=',',
    decimal='.'
).format_index(axis=1,formatter=str.upper)\
.relabel_index(["row1","row2"], axis=0)
```

```python
# 範例 DataFrame
data = {
    'A': [1234.5678, 9876.5432, 10000.00],
    'B': [0.1234, 0.5678, 0.9999],
    'C': ['蘋果', '香蕉', '櫻桃'],
    'D': pd.to_datetime(['2023-10-26', '2023-10-27', '2023-10-28']),
    'E': [np.nan, 1, 2]
}
df = pd.DataFrame(data)
df
```

```python
# 另一種使用 format() 方法格式化 DataFrame 的數值顯示（小數點後 2 位）
# subset=['A','B'] 只對 A 和 B 欄位進行格式化

styled_df1 = df.style.format("{:.2f}",subset=['A','B'])
styled_df1
```

```python
weather_df = pd.DataFrame(np.random.rand(10,2)*5,
                        index=pd.date_range(start="2021-01-01", periods=10),
                        columns=["Tokyo", "Taipei"])
weather_df
```

```python
weather_df.style\
.set_caption("Weather Conditions")\
.background_gradient(axis=None, vmin=1, vmax=5, cmap='BuPu')
```
